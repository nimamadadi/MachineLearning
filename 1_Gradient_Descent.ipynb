```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
data = pd.read_csv('score_updated.csv')
X = data['Hours']
Y = data.iloc[:, 1]
n = float(len(X)) 
plt.scatter(X, Y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
m_values = []  
loss_values = []  
min_loss = 1000000000
count=0
def f(x):
    return w*x

def MSE(Y,Y_pred):
    return (1/n) * sum(Y - Y_pred)**2

def d_MSE(Y,Y_pred):
    return (-2/n) * sum(X * (Y - Y_pred))  


for w in range(-10,30):
    count+=1
    Y_pred = f(X)
    Error = MSE(Y,Y_pred)
    if Error < min_loss:
        min_loss = Error
        Best_W = w
    loss_values.append(Error)
    m_values.append(w)

plt.plot(m_values, loss_values, color='r',label="Error Path", linewidth=2, markersize=12)
plt.xlabel('M')
plt.ylabel('Error')

w,L,epochs = 25,0.001,1000
W_values = [] 
loss1_values = [] 

for i in range(epochs):
    stop=w
    Y_pred = f(X)
    W_values.append(w) 
    Error = MSE(Y,Y_pred)
    loss1_values.append(Error)

    w -= L * d_MSE(Y,Y_pred) 

    if abs(stop-w) < 0.0001:
        break
    

plt.scatter(W_values, loss1_values, color='g', label="Gradient Descent Trajectory" )
plt.legend()
plt.grid(True)
print("W calulated by Gradient Descent is : ",w , "and intration= ",i)
print("W calulated by loss function is : ",Best_W, "and intration= ",count)

```
